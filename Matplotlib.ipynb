```python
##Matplotlib Labels and Title
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 169, 110, 147, 120, 175])
y = np.array([240, 250, 460, 270, 280, 290, 300, 540, 320, 230])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```
